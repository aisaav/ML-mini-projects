# Using PolyNomial Features to generate Polynomial inputs

Generating polunomial features allow us to use simpler ML models for our data. An example is using a simple linear regression model on polynomial data. We will be working with teh PolynomialFeatures object to help us preprocess our data.

First we will start by reating true polynomial data to then use polynomial features on.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
def true_polynomial_fn(x):
    
    return 5*x*x*x + 3*x*x + 2*x + 12 
```

```python
np.random.seed(0)

n_samples = 30
degree = 3

X = np.sort(np.random.rand(n_samples))
x = X.reshape(-1, 1)

y = true_polynomial_fn(X) + np.random.randn(n_samples) * 0.1
```

```python
y
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.plot(x, y , color='green', label="y")

plt.title('Y')
plt.legend()
```

```python
linear_regression_underfitted = LinearRegression().fit(x, y)

y_pred_underfitted = linear_regression_underfitted.predict(x)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.plot(x, y, color='green', label="y")
plt.plot(x, y_pred_underfitted, color='red', label="y_pred_underfitted")

plt.title('Y')
plt.legend()
```

As we can see since the data is not linear our linear regression estimator will be underfitted for our data, hence the variable name. Look how this improves by transforming x with scikit-learn's PolynomialFeatures

This will generate a feature matrix of all polynomial combinations of degree less than or equal to three

```python
polynomial_features = PolynomialFeatures(degree=3,
                                         include_bias=True)
```

```python
polynomial_features.fit_transform(x)
```

```python
linear_regression = LinearRegression()
pipeline = Pipeline([("polynomial_features", polynomial_features),
                     ("linear_regression", linear_regression)])
```

```python
pipeline.fit(x, y)
y_predicted = pipeline.predict(x)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.plot(x, y, color='green', label="y")

plt.plot(x, y_pred_underfitted, color='red', label="y_pred_underfitted")

plt.plot(x, y_predicted, color='blue', label="y_predicted")

plt.title('Y')
plt.legend()
```
