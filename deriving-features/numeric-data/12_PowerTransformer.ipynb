# Working with Chi-Squared Distributed Input Features

Power transformers are a family of transformers that are applied to the data when we want to make teh data more gausian like, power transformers are typically used with hertoscedastic data which has a non-constant variance to make it more normal-like. This notebook will focus on applying the PowerTransformer estimator object to convert data into gaussian distributions



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
```

```python
analyst_data = pd.read_csv('Datasets/Analyst_Forecasts.csv')
analyst_data.head()
```

```python
X = analyst_data[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8']]

y = analyst_data[['Votes']]
```

```python
X.head()
```

```python
y.describe()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.hist(analyst_data['Votes'], facecolor='yellow', edgecolor='red', bins=40)

plt.xlabel('# Votes')
plt.ylabel('Count')

plt.title('Histogram of # of Votes')

plt.show()
```

```python
analyst_data['Variability'] = analyst_data[['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8']].var(axis=1)

analyst_data.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.hist(analyst_data['Variability'], facecolor='purple', edgecolor='yellow', bins=40)

plt.xlabel('Variability')
plt.ylabel('Count')

plt.title('Histogram of Variability')

plt.show()
```

### Applying Power Transformers to get normal distributions

We call a simple linear regression to predict how popular an analysist is

```python
linear_regression = LinearRegression().fit(X, y)

linear_regression.score(X, y)
```

```python
y_predict = linear_regression.predict(X)
```

```python
results_df = pd.DataFrame({'y_actual': y['Votes'], 
                           'y_predicted': y_predict.reshape(1, -1)[0]})

results_df.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(y, y_predict, facecolor='blue', edgecolor='red', s=250)

plt.ylabel('# Votes (Predicted)')
plt.xlabel('# Votes (Actual)')

plt.title('Regression Results - Scatter')
```

```python
X_chi_sq = analyst_data[['Variability']]
X_chi_sq.head()
```

First we will use the Box-Cox PowerTransformer first, this requires data to be strictly positive and that input data is standardized to zero mean and unit variance by default

```python
power_transformer = PowerTransformer(method='box-cox')


X_transformed = power_transformer.fit(X_chi_sq).transform(X_chi_sq)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.hist(X_transformed, facecolor='cyan', edgecolor='red', bins=40)

plt.xlabel('Variability')
plt.ylabel('Count')

plt.title('Histogram of Variability')

plt.show()
```

```python
linear_regression = LinearRegression().fit(X_transformed, y)

linear_regression.score(X_transformed, y)
```

```python
y_predict_transformed = linear_regression.predict(X_transformed).reshape(1, -1)[0]

linear_regression.coef_
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(y, y_predict_transformed, facecolor='blue', edgecolor='red', s=250)

plt.ylabel('# Votes (Predicted)')
plt.xlabel('# Votes (Actual)')

plt.title('Regression Results - Scatter')
```

Next we will use the YeoJohnsom Transformer which works on positive and negative values

```python
power_transformer = PowerTransformer(method='yeo-johnson')

X_transformed_2 = power_transformer.fit(X_chi_sq).transform(X_chi_sq)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.hist(X_transformed_2, facecolor='cyan', edgecolor='red', bins=40)

plt.xlabel('Variability')
plt.ylabel('Count')

plt.title('Histogram of Variability')

plt.show()
```

```python
linear_regression = LinearRegression().fit(X_transformed_2, y)

linear_regression.score(X_transformed_2, y)
```

```python
y_predict_transformed_2 = linear_regression.predict(X_transformed_2).reshape(1, -1)[0]

linear_regression.coef_
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(y, y_predict_transformed_2, facecolor='blue', edgecolor='red', s=250)

plt.ylabel('# Votes (Predicted)')
plt.xlabel('# Votes (Actual)')

plt.title('Regression Results - Scatter')
```
